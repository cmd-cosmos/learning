```python
# any python object is a minimum of 16 bytes as --> Cpython interpretation has 8 bytes for ref count and 8 bytes pointing to the obj type
# ints are larger than 4 bytes due to com

import sys

x: int      = 100
size_x: int = sys.getsizeof(x)

print("sizeof x: {} bytes".format(size_x))
```

```python
# immutability at obj id level

print("id(x) when x = 100: {}".format(id(x)))
x += 1
print("id(x) when x = 101: {}".format(id(x)))
```

```python
# bit level introspection - endianness

n: int = 426
bin(n)
print(n.bit_length())
print(n.to_bytes(2, byteorder="big"))
```

```python
# float - IEEE 754

0.3 == 0.1 + 0.2
```

```python
import math

# nan is not equal to itself as per IEEE 754 rule
nan = float('nan')
print(math.isnan(nan))
print(math.nan == math.nan)
# floats are hashable so can be dict keys
print(hash(math.nan) == hash(math.nan)) # even though nan is undefined thus the hash is the same
```

```python
print(float('inf') > 1e308)
print(float('inf') > 1e309) # infinity overflow behavior
```
