```python
import sys

lst = []
print(" len       size")
for i in range(10):
    lst.append(i)
    print(f"{len(lst):3} : {sys.getsizeof(lst):4} bytes")
```

```python
import copy

l1 = [1, 2 ,3]
x = copy.copy(l1)
print("l1:  ",l1)
print("x:   ",x)
l1.append(10)
print("l1:  ",l1)
print("x:   ",x)

print(x is l1)
```

```python
test = [1,2,4,5,9]
y = copy.deepcopy(test)
print("test: ",test)
print("y   : ",y)
test.append(100)
print("test: ",test)
print("y   : ",y)

print(y is test)
```
