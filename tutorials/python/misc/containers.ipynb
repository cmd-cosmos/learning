```python
# lists --> dynamic array of object refs
# in cpython the list stores refs to pyobj instances

```

```python
import sys

lst = []
print(" len       size")
for i in range(10):
    lst.append(i)
    print(f"{len(lst):3} : {sys.getsizeof(lst):4} bytes")
```

```python
import copy

x = copy.copy(lst)
print("lst: ",lst)
print("x:   ",x)
x.append(10)
print("lst: ",lst)
print("x:   ",x)
```
