```python
### threading for I/O bound task concurrency

import threading
import time

def job(name, delay):
    time.sleep(delay)
    print(f"{name} : done after {delay} seconds")

threads = [
    threading.Thread(target=job, args=("A", 5)),
    threading.Thread(target=job, args=("B", 2)),
    threading.Thread(target=job, args=("C", 7)),
    threading.Thread(target=job, args=("D", 1))
]

for  t in threads:
    t.start()
for t in threads:
    t.join()

print("All threads finished execution.")
```

```python
### task

def task(n):
    time.sleep(1)
    print(f"task {n} done")
```

```python
# sequential execution

start = time.time()

res = [task(i) for i in range(5)]
print(res)
end = time.time()

print(f"time: {end - start}")
```

```python
import threading
import time

threads = []
start = time.time()

for i in range(5):
    t = threading.Thread(target=task, args=(i,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

end = time.time()
print(f"\ntime: {end - start}")
```
