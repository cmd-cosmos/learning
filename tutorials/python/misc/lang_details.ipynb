```python
# names vs values --> assignment and binding similarity
# assignment does not copy vals
# names bind to objects

a = [1,2,3]
b = a

b.append(4)

print("a = ", a)
print("b = ", b)

print("id a = ",id(a))
print("id b = ",id(b))

print(a is b)
```

```python
# pass by object reference

def mod(val, items):
    val += 1
    items.append("new")

num = 5  # immutable 
lst = [] # mutable

mod(num, lst)

print(num)
print(lst)

def deepMod(num):
    return num+1

# reassignment is the only way to update immutable vals
num = deepMod(num)
print(num)
```

```python
import time

class Timer:
    def __enter__(self):
        self.start = time.time()
        return self
    
    def __exit__(self, exc_type, exc, tb):
        print(f"Elapsed: {time.time() - self.start:.4f} s")

with Timer():
    total = sum(range(100_000_000))
    print(total)
```

```python
# runtime patching

class Service:
    def run(self):
        return "original"
    
s = Service()
print(s.run())

def newRun(self):
    return "patched"

Service.run = newRun # type: ignore

print(s.run())
```
