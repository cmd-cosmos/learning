```python
# for loop

for x in range(3):
    print(x, end=" ")
    x = 3

# range does not care about assignment inside the loop, it stops when the iterable is iterated over
```

```python
# vars are labels -> they can be robinded without affecting copies

a = 10
b = a
a = "batman"

print(b)
print(id(a))
print(id(b))
```

```python
# mutability

l = [1,2,3]
l2 = l
l.append(10)

print(l2)

t = (1,2,3)
t2 = t
t += (4,)
print(t2)
print(t)
```

```python
# default arg evaluation

def foo(x=[]):
    x.append(1)
    return x

print(foo())
print(foo())
print(foo())
print(foo.__defaults__) # created during function definition time and not call time
```

```python
# closures

def make_ctr():
    ct = 0
    def inc():
        nonlocal ct
        ct += 1
        return ct
    return inc

c1 = make_ctr()
c2 = make_ctr()

print(c1())
print(c1())
print(c1())
print(c1())
print()
print(c2())
print(c2())
print(c2())
```
