```python
# scalars are 0D tensors
# truly dense tensors   --> numpy.ndarray
# tensors with autodiff --> torch.tensor

import numpy as np
```

```python
vector = np.array([1, 2, 3])

matrix = np.array([
    [1, 2],
    [2, 1]
])

tensor3D = np.array([
    [[1, 2, 3]],  # 1x3 matrix
    [[4, 5, 6]],  # 1x3 matrix
    [[7, 8, 9]]   # 1x3 matrix
])

print(vector, '\n')
print(matrix, '\n')
print(tensor3D)
```

```python
print(vector.shape, '\n')
print(matrix.shape, '\n')
print(tensor3D.shape, '\n') # (3,1,3) --> 3 ==> 1x3 matrices
```

```python
print(tensor3D[0, 0, 1])
print(tensor3D[2, 0, :]) # print all elements from the 0th entry -> print all vals in the first row
```

```python
# tensor operations

a = np.array([[1, 2], [3, 4]])
b = np.array([[11, 12], [13, 14]])

print(a + b, '\n')
print(a - b, '\n')
```

```python
# matrix mult
a @ b
```

```python
# tensor reshaping

x = np.arange(10)
x_reshape = x.reshape(5,2)

print(x, '\n')
print(x_reshape, '\n')
```
