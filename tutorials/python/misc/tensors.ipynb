```python
# scalars are 0D tensors
# truly dense tensors   --> numpy.ndarray
# tensors with autodiff --> torch.tensor

import numpy as np
```

```python
vector = np.array([1, 2, 3])

matrix = np.array([
    [1, 2],
    [2, 1]
])

tensor3D = np.array([
    [[1, 2, 3]],  # 1x3 matrix
    [[1, 2, 3]],  # 1x3 matrix
    [[1, 2, 3]]   # 1x3 matrix
])

print(vector, '\n')
print(matrix, '\n')
print(tensor3D)
```

```python
print(vector.shape, '\n')
print(matrix.shape, '\n')
print(tensor3D.shape, '\n') # (3,1,3) --> 3 ==> 1x3 matrices
```
