```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
```

```python
for row in matrix:
    print(row)
```

```python
def print_matrix(matrix):
    for row in matrix:
        for col_val in row:
            print(col_val, end=" ")
        print()
print_matrix(matrix=matrix)
```

```python
# modif a val at a certain coord in the matrix

# (m,n) -> (3,3) -> mod val at 3rd row and 3rd col
# 1 2 3
# 4 5 6
# 7 8 10000 <---- mod

for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        if i == j == 2:
            matrix[i][j] = 10000
        continue

print_matrix(matrix=matrix)
```
