```python
## Strings in detail
# strings are unicode by default
# internally are PyUnicodeObject objects in CPython

type("batman")
```

```python
s = "batman"

print(id(s))
print(id(s + " is here")) # temp modification
print(s) # immutable
```

```python
print(str(100))
print(str(3.14156))
print(str(None))
print(type(str(b"bytes")))
print(type(b"I am Batman"))
```
