```python
# factorials

def factorial(n):
    if n == 0 or n == 1:
        return 1
    res = 1
    for i in range(2, n+1):
        res *= i
    return res

print(factorial(0))
print(factorial(1))
print(factorial(5))
print(factorial(7))
```

```python
import itertools

def all_permutations(ele):
    return list(itertools.permutations(ele))

nums = [1,2,3]
all_permutations(nums)
```

```python
def all_combinations(ele, k):
    return list(itertools.combinations(ele, k))

all_combinations([1,2,3,4], 2)
```

```python
def power_set(ele):
    res = []
    for r in range(len(ele)+1):
        res.extend(itertools.combinations(ele, r))
    return res

power_set([1,2,3])
```

```python
# catalan numbers
import math

def catalan(n):
    return math.comb(2*n,n) // (n+1)

[catalan(i) for i in range(6)]
```
