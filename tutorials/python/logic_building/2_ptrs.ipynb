```python
# 2 pointers algo
# optimized iteration method
# for value based ptr movement, sorting is necessary

# array

array = [i for i in range(1, 100, 10)]
array 
```

```python
# brute force to solve a 2 sum pattern problem
# in the array find nums that add to 112

def brute_force(arr, target):
    n = len(arr)
    for i in range(n):
        print(f"At: i = {i}", end=",")
        for j in range(i+1, n):
            print(f"-> j = {j}", end="\n")
            if arr[i] + arr[j] == target:
                return i,j
    return None

brute_force(array, 112)
```

```python
brute_force(arr=array, target=152)
```

```python
# 2 pointers

left = 0
right = len(array) - 1

while left < right:
    print(f"left ptr at : {left:5} | val: {array[left]}")
    print(f"right ptr at: {right:5} | val: {array[right]}\n")
    left += 1
    right -= 1
```

```python
# 2 pointers for value based pointer movement requires sorting

array2 = [12, 19, 22, 36, 54, 81, 7, 91, 63, 121, 41]

def two_ptr(arr, target):
    arr.sort()
    left = 0
    right = len(arr) - 1
    while left < right:
        curr_sum = arr[left] + arr[right] 
        if curr_sum == target:
            print("vals: {}, {}".format(arr[left], arr[right]))
            print("indices: {}, {}".format(left, right))
            break
        elif curr_sum < target:
            left += 1
        else:
            right -= 1
    else:
        print("not found")

two_ptr(arr=array2, target=145)
```
