```python
# 2 pointers algo
# optimized iteration method

# array

array = [i for i in range(1, 100, 10)]
array 
```

```python
# brute force to solve a 2 sum pattern problem
# in the array find nums that add to 92

def brute_force(arr, target):
    n = len(arr)
    for i in range(n):
        print(f"At: i = {i}", end=",")
        for j in range(i+1, n):
            print(f"j = {j}", end="\n")
            if arr[i] + arr[j] == target:
                return i,j
    return None

brute_force(array, 92)
```
