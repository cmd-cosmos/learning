```python
class Point:
    def __init__(self, x, y) -> None:
        self.x = x
        self.y = y
    
    # called when object is printed - meant for userss
    def __str__(self) -> str:
        return f"Point({self.x}, {self.y})"
    
    #  called when object name written in repl without passing it to print()- meant for programmers
    def __repr__(self) -> str:
        return f"Point(x={self.x}, y={self.y})"

p2 = Point(2,2) 
print(p2) # calls __str__
p2        # calls __repr__
```

```python
# arithmetic dunder  methods

class Point:
    def __init__(self, x, y) -> None:
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Point(self.x - other.x, self.y - other.y)
    
    def __repr__(self) -> str:
        return f"Point(x={self.x}, y={self.y})"

p3 = Point(12,33)
p4 = Point(10,17)

print(p3 + p4)
print(p3 - p4)
```

```python
# comparision dunder methods

class Point:
    def __init__(self, x, y) -> None:
        self.x = x
        self.y = y

    def __eq__(self, other) -> bool:
        return self.x == other.x and self.y == other.y
    
    def __lt__(self, other):
        return (self.x**2 + self.y**2) < (other.x**2 + other.y**2)

pt1 = Point(1,1)
pt2 = Point(3,4)
pt3 = Point(1,1)

print(pt1 == pt2)
print(pt1 == pt3)
print(pt1 < pt2) # origin proximity comparision
```

```python
class Circle:
    pi = 3.14159

    def __init__(self, radius) -> None:
        self.radius = radius
    
    @classmethod
    def from_diam(cls, diameter):
        # create obj of same class using diameter instead of radius
        return cls(diameter / 2)
    
    @staticmethod
    def area(radius):
        return Circle.pi * radius**2

c1 = Circle(10)
c2 = Circle.from_diam(100)

print(c1.radius, c2.radius)
print(Circle.area(3))
```

```python
class DataStruct:
    def __init__(self, items) -> None:
        self.items = items
    
    def __len__(self):
        return len(self.items)
    
    def __getitem__(self, index):
        return self.items[index]

ds = DataStruct([1,2,3,4,5])
print(len(ds))
print(ds[3])
```

```python
class Vector:
    def __init__(self, x, y) -> None:
        self.x = x 
        self.y = y

    def __mul__(self, scalar):
        # allows vec * scalar but not scalar * vec
        return Vector(self.x * scalar, self.y * scalar)
    
    def __rmul__(self, scalar):
        # allows scalar * vec operations
        return self.__mul__(scalar=scalar)
    
    def __matmul__(self, other):
        # overrides the @ operator for matrix products
        return self.x * other.x + self.y * other.y
 
v1 = Vector(1,2)
s = 10

print((v1*s).x)
print((v1*s).y)
# print((s*v1).x) <--- scalar * vector not allowed by dunder mul override
print((12*v1).x)
print((12*v1).y)

v2 = Vector(0,0)
print(v1 @ v2)
```
