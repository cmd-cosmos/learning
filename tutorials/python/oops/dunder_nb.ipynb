```python
class Point:
    def __init__(self, x, y) -> None:
        self.x = x
        self.y = y
    
    # called when object is printed - meant for userss
    def __str__(self) -> str:
        return f"Point({self.x}, {self.y})"
    
    #  called when object name written in repl without passing it to print()- meant for programmers
    def __repr__(self) -> str:
        return f"Point(x={self.x}, y={self.y})"

p2 = Point(2,2) 
print(p2) # calls __str__
p2        # calls __repr__
```
