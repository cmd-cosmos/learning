```python
# hashing algorithms
# convert input data into a fixed size string of bytes

import hashlib
```

```python
# MD5 - not secure for cryptography
#       output length = 128 bits / 32 hex characters
#       useful for checksums and file integrity

text = "batman"

md5_hash = hashlib.md5(text.encode())
md5_hash.hexdigest()
```
