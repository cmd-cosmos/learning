```python
import sqlite3
import pandas as pd
```

```python
# in memory db
conn = sqlite3.connect(":memory:")
cur = conn.cursor()
```

```python
cur.execute("""
CREATE TABLE customers (id INTEGER PRIMARY KEY, name TEXT, age INTEGER, amount REAL)
""")
```

```python
conn.commit()
```

```python
cur.execute("""
INSERT INTO customers (name, age, amount)
VALUES ('Batman', 22, 1000)
""")
conn.commit()
```

```python
# insert multiple rows
# ? is used to prevent SQL injections, it acts as a placeholder

data = [
    ('Joker', 30, 300),
    ('Alfred', 75, 2000)
]

cur.executemany("""
INSERT INTO customers (name, age, amount) VALUES (?, ?, ?)
""", data)

conn.commit()
```

```python
# query data

cur.execute("SELECT * from customers")
rows = cur.fetchall()

for row in rows:
    print(row)
```

```python
# filtering

cur.execute("SELECT name, amount FROM customers WHERE  amount > 300")
print(cur.fetchall())
```

```python
# pandas and sql interaction

df = pd.read_sql("SELECT * FROM customers", conn)
df
```

```python
conn.close()
```
