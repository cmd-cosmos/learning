```python
# range based creation

range_arr = np.arange(0, 10, 2)
print(range_arr, end="\n\n")

linspace = np.linspace(0, 1, 5)
print(linspace) # equally spaced nums
```

```python
arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(arr.shape, end="\n\n") # rows x cols
print(arr.ndim, end="\n\n")  # num of axes of an arr -> 2D matrix
print(arr.size, end="\n\n")  # total eles in an arr
print(arr.dtype, end="\n\n") # data type of eles in matrix
```

```python
# indexing and listing

lst = np.array([1, 2, 3, 4, 5, 6])

print(lst[0], end="\n\n")
print(lst[-1], end="\n\n")
print(lst[:3], end="\n\n")
```

```python
# 2D indexing

matrix = np.array([
    [1,2,3],
    [10,20,30],
    [11,12,13]
])

print(matrix[1, 2], end="\n\n") # row 1 col 2
print(matrix[:, 1], end="\n\n") # all row vals from col 1
```

```python
# vectorized operations

# no loops required --> any linear operation run without loops on numpy arrs

vec = np.array([1,2,3,4,5])

print(vec + 100, end="\n\n")
print(vec * 100, end="\n\n")
print(vec ** 2, end="\n\n")
```

```python
print(np.sqrt(vec), end="\n\n")
print(np.mean(vec), end="\n\n")
print(np.median(vec), end="\n\n")
print(np.std(vec), end="\n\n")
print(np.min(vec), np.max(vec), end="\n\n")
```

```python
# boolean masking

test = np.array([1, 2, 3, 4, 5])

mask = test > 2.5

print(test[mask])

print(test[test % 2 == 0])
```

```python
# reshaping and flattening arrs

arr2 = np.arange(9)
print(arr2)
print(arr2.shape, end="\n\n")

res = arr2.reshape(3, 3)
print(res, end="\n\n")

flat = res.flatten()
print(flat)
```

```python
# stacking arrays -> vstack and hstack

t1 = np.array([1,2,3])
t2 = np.array([4,5,6])

print(np.vstack((t1, t2)), end="\n\n") # vertical stacking
print(np.hstack((t1, t2)), end="\n\n") # horizontal stacking
```

```python
# splitting

t3 = np.array([1,2,3,4,5,6,7,8])
np.split(t3, 4) # split t3 into 4 arrays
```

```python
# random module 

np.random.seed(42)

rand_nums = np.random.rand(5)
print(rand_nums)

randints = np.random.randint(1, 20, size=5)
print(randints)
```

```python
t4 = np.array([1, 2, np.nan, 4, np.nan])

print(np.isnan(t4))
print(np.nanmean(t4)) # calc mean of arr and ignore the nan vals
```
