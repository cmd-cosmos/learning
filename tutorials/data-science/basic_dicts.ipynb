```python
## dicts

d = {
    1 : "batman",
    2 : "joker",
    3 : "alfred",
    4 : "scarecrow"
}

# key based access
d[1]
```

```python
# other dict creation methods

d2 = dict(a=1, b=2) # this would not allow int keys
print(d2)

d3 = dict([(1, "a"), (2, "b")]) # tuple based dict creation
print(d3)
```

```python
# unpacking items
print(d.items())

for k, v in d.items():
    print(k,":",v)
```

```python
## creating a list of keys
print([i for i in d.keys()])
print([i for i in d.values()])
```

```python
### access

# d["batman"] <--- keyerror
d.get("batman") # no errors raised
d.get("batman", 0) # default val returned in case of missing key

print(d)
```

```python
## adding and updating

d["PI"] = 3.1415
d[4] = "RIDDLER"

print(d)
```

```python
d.popitem() # remove last inserted item
d
```

```python
print(d2)
d2.pop("a")
print(d2)
```

```python
print(d3)
del d3[2]
print(d3)
```

```python
# dict comprehensions

sq = {x : x**2 for x in range(10)}
print(sq)

evens = {x : x for x in range(10) if x % 2 == 0}
print(evens)
```

```python
# frequency counter for fast lookups

line = "banana"
freq = {}

for char in line:
    freq[char] = freq.get(char, 0) + 1

freq
```
