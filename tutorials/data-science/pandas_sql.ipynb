```python
# pandas equivalent of SQL statements

import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "id" : [1,2,3,4,5],
    "name" : ["Batman", "Alfred", "Robin", "Joker", "Scarecrow"],
    "language" : ["C", "C++", "Python", "C", "ASM"],
    "salary" : [10000000, 100000, 950000, 65000000, 80000],
    "year" : [2000, 1997, 2020, 2000, 2003] 
})
```

```python
# SQL: SELECT * FROM DATASET

df
```

```python
# SQL: SELECT x,y FROM DATASET

df[["id", "language"]]
```

```python
# where filtering
# SQL: SELECT * FROM dataset WHERE name = "batman"

df[df.language == "C"]
```

```python
# multiple condition filtering
# SQL: WHERE id = 1 and dept = "IT"

df[(df.language == "C") & (df.id == 1)]
```
