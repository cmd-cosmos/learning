```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# synthetic data

x = np.linspace(0, 10, 100)
y = np.sin(x)

print(x, end="\n\n")
print(y, end="\n\n")
```

```python
# basic plot

plt.plot(x, y)
plt.title("basic plot: sin(x)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
```

```python
# scatter plot

plt.scatter(x, y, color="black", marker="x")
plt.title("scatter plt: sin(x)")
plt.show()
```

```python
# bar plot

cat  = ['A', 'B', 'C', 'D']
vals = [8, 25, 34, 3]

plt.bar(cat, vals, color='black')
```

```python
# histograms

data = np.random.randn(1000)
plt.hist(data, bins=10, align="mid", edgecolor="black")
print(data[0:100])
plt.show()
```

```python
# subplots

fig, axis = plt.subplots(2,1, figsize=(10,8))

axis[0].plot(x, np.sin(x))
axis[0].set_title('sin(x)')
axis[1].plot(x, np.cos(x))
axis[1].set_title('cosine(x)')
```
