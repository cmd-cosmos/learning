```python
import time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use('seaborn-v0_8-darkgrid')
```

```python
# synthetic data

x = np.linspace(0, 10, 100)
y = np.sin(x)

print(x, end="\n\n")
print(y, end="\n\n")
```

```python
# basic plot

plt.plot(x, y)
plt.title("basic plot: sin(x)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
```

```python
# scatter plot

plt.scatter(x, y, color="black", marker="x")
plt.title("scatter plt: sin(x)")
plt.show()
```

```python
# bar plot

cat  = ['A', 'B', 'C', 'D']
vals = [8, 25, 34, 3]

plt.bar(cat, vals, color='black')
```

```python
# histograms

data = np.random.randn(1000)
plt.hist(data, bins=10, align="mid", edgecolor="black")
print(data[0:100])
plt.show()
```

```python
# subplots

fig, axis = plt.subplots(2,1, figsize=(10,8))

axis[0].plot(x, np.sin(x))
axis[0].set_title('sin(x)')
axis[1].plot(x, np.cos(x))
axis[1].set_title('cosine(x)')
```

```python
# pie chart

flights  = [20, 30, 12, 38]
labels   = ["proton", "falcon-heavy", "starship", "falcon 9"]

plt.pie(flights, labels=labels)
plt.show()
```

```python
speed = [i for i in range(0, 100, 10)]
time_steps = [t for t in range(0, 10)]

print(speed)
print(time_steps)

plt.ion()

x,y = [], []

for i in range(len(time_steps)):
    x.append(time_steps[i])
    y.append(time_steps[i])

    plt.plot(x, y)
    plt.pause(1)

plt.ioff()
plt.show()
```
