```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# synthetic data

x = np.linspace(0, 10, 100)
y = np.sin(x)

print(x, end="\n\n")
print(y, end="\n\n")
```

```python
# basic plot

plt.plot(x, y)
plt.title("basic plot: sin(x)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
```
