```python
import numpy as np
```

```python
# standardization

x = np.array([
    [10 ,200],
    [60 ,180],
    [100 ,2000]
])

x_std = (x - x.mean(axis=0)) / x.std(axis=0)
print(x_std, end="\n\n")
print(x_std.mean(), end="\n\n") # almost 0 mean
print(x_std.std(), end="\n\n")  # std dev = 1
```

```python
# train validation split - manual

np.random.seed(42)

a = np.random.rand(100, 5)
b = np.random.randint(0, 2, size=100)

indices = np.random.permutation(len(a))
train_size = int(0.7 * len(a))

train_idx, val_idx = indices[:train_size], indices[train_size:]

a_train, a_val = a[train_idx], a[val_idx]
b_train, b_val = b[train_idx], b[val_idx]

print(a_train, end="\n\n")
print(b_train, end="\n\n")
```
