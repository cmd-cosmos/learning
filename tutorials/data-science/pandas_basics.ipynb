```python
import pandas as pd

df = pd.read_csv("dirty_cafe_sales.csv")
df.head()
```

```python
# basic data descriptions
# info and describe --> info returns structure and nulls whereas describe returns stats

print(df.info(), end="\n\n")
print(df.describe(), end="\n\n")
```

```python
# more inspection

print(df.tail(), end="\n\n")
print(df.shape, end="\n\n")
print(df.columns, end="\n\n")
print(df.dtypes, end="\n\n")
```

```python
# col selection

print(df["Payment Method"][:10], end="\n\n")
print(df[["Item", "Location"]])
```

```python
# adding or modifying cols

df["Test"] = df["Quantity"] + " world"
print(df.Test, end="\n\n")
print(df["Test"])
```

```python
# row selection --> loc vs iloc
# loc  --> label based row selection
# iloc --> position based row selection

print(df.iloc[0], end="\n\n") # row 0
print(df.loc[0, "Item"]) # val for row 0 col Item
```

```python
# filtering rows using bool idx

print(df[df["Total Spent"] > "9"], end="\n\n\n")
print(df[(df["Total Spent"] > "5") & (df["Location"] == "Takeaway")])
```

```python
df.query("Item == 'Cake' and Quantity > '3'")
```
