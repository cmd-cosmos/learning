```python
import warnings
import numpy as np

warnings.filterwarnings("ignore")

print(np.__version__)
```

```python
# arrays

arr = np.array([1,2,3]) # rank = 1
print(type(arr))
print(arr.shape)
print(arr[0], arr[2])
arr[0] = np.zeros(1)
print(arr)
```

```python
nums = np.array([[1,2,3],
                 [4,5,6]])

print(f"dimensions: {nums.ndim}")
print(f"size: {nums.size}")
print(f"dtype: {nums.dtype}")
print(f"itemsize: {nums.itemsize}")
```

```python
print(np.array([1,2,3], dtype=float))
print(np.array([1,2,3], dtype=np.int32))
print(np.array([1,2,3], dtype=np.bool))
```

```python
# rank 2 array

b = np.array([[1,2], [3,4]])
print(b)
print(b.shape)
print(type(b))

for rows in b:
    print(rows)

print(b[0, 1]) # value in the 0th row and 1st col
```

```python
# other methods of array creation

t1 = np.zeros((3,3))
print(t1, end="\n\n")

t2 = np.ones((2,2))
print(t2, end="\n\n")

t3 = np.full((4,4), 11)
print(t3, end="\n\n")

# identity matrix
t4 = np.eye(3) 
print(t4, end="\n\n")

t5 = np.random.random((2,2))
print(t5)
```
