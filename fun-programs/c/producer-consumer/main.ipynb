```python
import numpy as np
import matplotlib.pyplot as plt 

sims = 1_000_000

A = np.random.uniform(1, 4, sims)
B = np.random.uniform(2, 6, sims)

duration = A + B
```

```python
plt.figure()
plt.hist(duration, density=True, edgecolor='black')
plt.axvline(9, color='r')
plt.show()
print((duration > 9).sum() / sims)
```
