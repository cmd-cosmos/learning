```python
# single perceptron for binary classification

import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
outputs = np.array([[0], [0], [0], [1]])

weights = np.random.random((2,1))
bias = np.random.random((1))
```

```python
learning_rate = 0.1
epochs = 10_000

for epoch in range(epochs):
    lin_out = np.dot(inputs, weights) + bias
    out = sigmoid(lin_out)

    err = outputs - out

    d_error = err * sigmoid_derivative(out)
    weights += np.dot(inputs.T, d_error) * learning_rate
    bias += np.sum(d_error) * learning_rate

    if epoch % 1000 == 0:
        print(f"Epoch {epoch:4.0f} | Error: {np.mean(err)}")
```
