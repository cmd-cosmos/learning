```python
from transformers import pipeline

llm = pipeline(
    "text-generation",
    model="HuggingFaceTB/SmolLM2-1.7B-Instruct",
    max_new_tokens=256,
)
```

```python
import datetime

def get_time():
    return str(datetime.datetime.now())

def calculator(expression):
    return str(eval(expression))
```

```python
TOOLS = {
    "get_time" : get_time,
    "calculator" : calculator
}
```

```python
SYSTEM_PROMPT = """
You are an AI agent.

Available tools:
- get_time()
- calculator(expression)

Rules:
1. Never calculate answers yourself.
2. Never guess tool outputs.
3. To use a tool, output exactly:

ACTION: tool_name
INPUT: tool_input

4. After receiving an OBSERVATION, continue reasoning.
5. When you have enough information, output:

FINAL: answer
"""
```

```python
# llm wrapper function
def ask_llm(prompt):
    output = llm(
        prompt,
        max_new_tokens=128,
        do_sample=False
        )    
    print(output)
    return output[0]["generated_text"]
```

```python
def run_agent(question):
    history = f"{SYSTEM_PROMPT}\n\nUser: {question}\n"
    for step in range(5):
        response = ask_llm(history)
        print(f"\nSTEP {step}")
        print(response)
        if "FINAL:" in response:
            return response
        if "ACTION:" in response:
            lines = response.split("\n")
            action = None
            tool_input = ""
            for line in lines:
                if line.startswith("ACTION:"):
                    action = line.replace("ACTION:", "").strip()
                if line.startswith("INPUT:"):
                    tool_input = line.replace("INPUT:", "").strip()
            result = TOOLS[action](tool_input) \
                if tool_input else TOOLS[action]()

            history += f"""
Assistant: {response}

OBSERVATION:
{result}
"""

    return "Agent stopped."
```

```python
run_agent(
    "What is 456 * 789 and what time is it?"
)
```
