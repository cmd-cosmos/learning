```python
import os
from huggingface_hub import HfApi, HfFolder, login, hf_hub_download, snapshot_download
from dotenv import load_dotenv

load_dotenv()
login(token=os.getenv("HF_KEY"))
```

```python
api = HfApi()

models = api.list_models(limit=10)
for model in models:
    print(model.modelId)
```

```python
text_models = api.list_models(filter="text-generation", sort="downloads", limit=10)
for tm in text_models:
    print(tm.modelId)
```
