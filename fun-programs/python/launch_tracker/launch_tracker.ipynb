```python
import requests
import json

base_url = f"https://fdo.rocketlaunch.live/json/launches/next/5"
```

```python
try:
    response = requests.get(base_url)
    print(response.status_code)
    data = response.json()
    data_json_formatted = json.dumps(data, indent=4)
    print(data_json_formatted)
except Warning:
    print(f"Error fetching data - {response.status_code}")
```

```python
data_results = data["result"]
data_results
```

```python
_id = "ID"
_mission_name = "Mission Name"
_agency = "Agency"
_vehicle = "Vehicle"
_t0 = "T-0 Time"

print(f"{_id.ljust(5)} | {_mission_name.ljust(30)} | {_agency.ljust(15)} | {_vehicle.ljust(15)} | {_t0.ljust(25)}")

print("-" * 110)
for items in data_results:
    print(f"{str(items["id"]).ljust(5)} | {items["name"].ljust(30)} | {items["provider"]["name"].ljust(15)} | {items["vehicle"]["name"].ljust(15)} | {str(items["t0"]).ljust(25) if items["t0"] is not None else str(items["modified"].ljust(45))}")
```
