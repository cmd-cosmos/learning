```python
import time
import cProfile
from IPython.display import SVG
```

```python
def slow_square(n):
    time.sleep(0.01)
    return n * n

def compute_series(n):
    total = 0
    for i in range(n):
        total += slow_square(i)
    return total

def wrapper():
    compute_series(20)
    compute_series(30)

wrapper()
```

```python
cProfile.run("wrapper()", "output.prof")
```
