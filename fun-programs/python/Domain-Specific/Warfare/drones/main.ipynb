```python
import time
import random
import numpy
from typing import Any, List, Optional, Tuple
```

```python
class Drone:
    def __init__(self, id: int, x: float, y: float, speed: float = 0, target:Any = None) -> None:
        self.id      : int   = id
        self.x       : float = x
        self.y       : float = y
        self.speed   : float = speed
        self.target  : Any   = target
        self.status  : bool  = True
        self.battery : float = 100

    def position(self) -> Tuple:
        roundedX = f"{self.x:06.2f}"
        roundedY = f"{self.y:06.2f}"
        return (roundedX, roundedY)
```

```python
class DroneSwarm:
    def __init__(self, swarmSize: int, targetPosition: List) -> None:
        self.drones : List = []
        self.targetPosition : List = targetPosition
        self.createSwarm(swarmSize)
    
    def createSwarm(self, swarmSize: int):
        for i in range(swarmSize):
            x = random.uniform(0, 100)
            y = random.uniform(0, 100)

            drone = Drone(id=i, x=x, y=y, speed=random.uniform(1,5), target=self.targetPosition)

            self.drones.append(drone)
```

```python
swarm = DroneSwarm(15, [50, 50])

for drone in swarm.drones:
    print(f"DRONE ID {drone.id}".ljust(11), f"| POS: {drone.position()}".ljust(15), f"| TARGET: {drone.target}")
```
