```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
```

```python
EARTH_RAD = 6371

def generate_sat(num_sat=4, altitude=20000):
    satellites = []
    for i in range(num_sat):
        angle = i * (2 * np.pi / num_sat) 
        x = (EARTH_RAD + altitude) * np.cos(angle)
        y = (EARTH_RAD + altitude) * np.sing(angle)
        z = 0

        satellites.append([x,y,z])
    
    return np.array(satellites)
```
