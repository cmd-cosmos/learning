```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
```

```python
EARTH_RAD = 6378

def generate_sat(num_sat=4, altitude=20000):
    satellites = []
    for i in range(num_sat):
        angle = i * (2 * np.pi / num_sat) 
        x = (EARTH_RAD + altitude) * np.cos(angle)
        y = (EARTH_RAD + altitude) * np.sin(angle)
        z = 0

        satellites.append([x,y,z])
    
    return np.array(satellites)

sat_positions = generate_sat(6)
sat_positions
```

```python
# true receiver position
true_position = np.array([1000, 2000, 300])

print(f"Receiver Coordinates: {true_position}")
```

```python
def compute_distances(receiver_pos, satellites):
    distances = []

    for sat in satellites:
        dist = np.linalg.norm(receiver_pos - sat)
        distances.append(dist)
    
    return np.array(distances)

true_distances = compute_distances(true_position, sat_positions)
true_distances
```

```python
def add_noise(distances, noise_level=0.01):
    noise = np.random.normal(0, noise_level, size=len(distances))
    return distances + noise

measured_distances = add_noise(true_distances)
measured_distances
```

```python
def residuals(x, satellites, measured_distances):
    predicted = np.linalg.norm(satellites - x, axis=1)
    return predicted - measured_distances

init_guess = np.array([0,0,0])
result = least_squares(residuals, init_guess, args=(sat_positions, measured_distances))

estimated_pos = result.x
print(estimated_pos)
```

```python
print(f"True Position: {true_position}")
print(f"Est. Position: {estimated_pos}")
print(f"Error (km): {np.linalg.norm(true_position - estimated_pos)}")
```
