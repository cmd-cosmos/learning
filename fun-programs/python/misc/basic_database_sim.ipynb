```python
class Table:
    def __init__(self, name, col) -> None:
        self.name = name
        self.col = col
        self.row = []
    
    def insert(self, **kwargs):
        row = {}
        for col in self.col:
            row[col] = kwargs.get(col, None)
        self.row.append(row)
    
    def select(self, col=None, where=None):
        col = col or self.col
        res = []
        for row in self.row:
            if where is None or where(row):
                res.append({c : row[c] for c in col})
        return res
```

```python
class Database:
    def __init__(self) -> None:
        self.tables = {}

    def create_table(self, name, cols):
        self.tables[name] = Table(name, cols)
    
    def get_tables(self, name):
        return self.tables.get(name)
```

```python
db = Database()

db.create_table("users", ["id", "name", "age"])
users = db.get_tables("users")
```
