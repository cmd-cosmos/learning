```python
import random

class SensorError(Exception):
    pass

def read_altimeter():
    if random.random() < 0.2:
        raise SensorError("Altimeter Reading Failed.")
    return random.uniform(100.0, 500.0)

try:
    alt_reading = read_altimeter()
    print(f"Altitude: {alt_reading:.3f} meters")
except SensorError as e:
    print(f"Warning: {e}")
```
