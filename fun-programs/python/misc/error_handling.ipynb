```python
import random

class SensorError(Exception):
    pass

def read_altimeter():
    if random.random() < 0.2:
        raise SensorError("Altimeter Reading Failed.")
    return random.uniform(100.0, 500.0)

try:
    alt_reading = read_altimeter()
    print(f"Altitude: {alt_reading:.3f} meters")
except SensorError as e:
    print(f"Warning: {e}")
```

```python
class MotorError(Exception):
    pass

def move_arm(position):
    if not (0 <= position <= 180):
        raise ValueError(f"Target position {position} degrees out of range | Envelope Range: (0-180 degrees)")
    if random.random() < 0.1:
        raise MotorError("Motor stalling ==> check power levels")
    print(f"Robot arm moved to: {position:3} degrees")

positions = [30, 100, 80, 45, 150, 200]

for pos in positions:
    try:
        move_arm(pos)
    except ValueError as ve:
        print(f"Input Error: {ve}")
    except MotorError as me:
        print(f"Hardware Error: {me}")
        try:
            move_arm(pos)
        except MotorError:
            print("Retry failure: moving to home position (0,0)")
            move_arm(0)
    
```
