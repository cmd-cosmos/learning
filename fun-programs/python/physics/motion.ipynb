```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear(u, a, t):
    '''
    u: init velocity (m/s)
    a: acc (m/s^2)
    t: time (s) 

    return velocity for an object in motion at time t
    '''
    return u + a*t

v = linear(10, 2, 100)
print(f"v = {v} m/s")
```

```python
# velocity time graph for linear motion
vel = []
t_s = []
for i in range(100):
    v_calc = linear(10, 2, i)
    vel.append(v_calc)
    t_s.append(i)

# print(vel)
# print(t_s)
plt.plot(t_s, vel)
```

```python
# second equation of motion

def second_eqn(v0, t, a):
    '''
    v0: initial velocity (m/s)
    t : time (s)
    a : accelaration (m/s^2)
    '''
    return v0*t + 0.5*a*t**2

disp = second_eqn(v0=0, t=6, a=2)
print(f"displacement = {disp} m")
```

```python
disp_pts = []
t_stamps = []

for i in range(100):
    disp_calc = second_eqn(0, i, 2)
    disp_pts.append(disp_calc)
    t_stamps.append(i)

print(disp_pts)
print(t_stamps)

plt.plot(t_stamps, disp_pts)
```

```python
def simple_harmonic(A, omega, t, phi):
    '''
    A     : amplitude (m)
    omega : angular frequency (rad / s)
    t     : time (s)
    phi   : phase angle (rad)
    '''
    return float(A * np.cos(omega*t + phi))

shm = simple_harmonic(0.10, 20, 0.05, 0)
print(f"X = {shm:0.5f} m")
```
